### 空值处理
根据数据分布情况（空值多不多），采用不同的处理方法：
* 某条记录全为空值——删除该记录；
* 某列数据空值较多——删除该列；
* 某列数据空值一半左右——使用均值填充；
* 某列数据空值三分之一左右——使用0填充；
* 某列数据空值极少——删除该记录
